```
# import libraries
import matplotlib.pyplot as plt
import itertools as it
import numpy as np
```

```
def bubbleSort(v):
  steps = 0
  chain = v[:]
  for i in range(len(chain)-1):
    for j in reversed(range(i+1,len(chain))):
      steps += 1
      if (chain[j] < chain[j-1]):
        swap(chain,j,j-1)
        steps+=1
  return steps, chain

def optimizedBubbleSort(v):
  steps = 0
  chain = v[:]
  for i in range(len(chain)-1):
    update = False
    for j in reversed(range(i+1,len(chain))):
      steps += 1
      if (chain[j] < chain[j-1]):
        swap(chain,j,j-1)
        steps+=1
        update = True
    if (not update):
      break
  return steps, chain

def swap(v,i,j):
  t = v[i]
  v[i] = v[j]
  v[j] = t
```

```
def plotBSort(n,opc=True):
  ''' 
    Grafica los pasos que se demora el algoritmo bubble sort.
    Si opc: True ejecuta bubble sort optimizado
    Si opc: False ejecuta bubble sort
    n: es el tamaÃ±o de la lista a ordenar, dado la RAM, no corre con n>10
    n: entre 0 y 10
  '''
  v = [i+1 for i in range(n)] # Vector con n elementos
  pobj = it.permutations(v)   # Se calculan las n! permutaciones
  permlist = list(pobj)
  xhist = []
  if (not opc):               # Si opc == False, ejecuta bubbleSort
    for p in permlist:
      pl = list(p)
      t, po = bubbleSort(pl)
      xhist.append(t)
  else:                       # Si opc == True, ejecuta optimizedBubbleSort
    for p in permlist:
      pl = list(p)
      t, po = optimizedBubbleSort(pl)
      xhist.append(t)
 
  titulo = "Bubble Sort time distribution calculated for N={0}".format(n)
  plt.hist(xhist, density=True)
  plt.title(titulo)
  plt.ylabel("density")
  plt.xlabel("time complexity")
  plt.show()

  for i in range(min(xhist), max(xhist)+1):
    porcentaje = xhist.count(i)/len(xhist)*100
    print("P({0}) = {1} %".format(i,porcentaje))
  print("\nAvg:",sum(xhist)/len(xhist))

  
```

Ejemplo con n=5

```
plotBSort(5)
```

```
import time
```

```
def fibR(n): # Recursive Fibonacci
    if n==0 :
        return 0
    if n==1 :
        return 1
    return fibR(n-1)+fibR(n-2)

def fib(n): # Iterative Fibonacci
    if n==0 :
        return 0
    a, b = 0, 1
    for i in range(1, n):
        a, b = b, a+b
    return b
```

```
time_list = []
n_list = []
n=50
tstart = time.clock()
for i in range(n):
  ti = time.clock()
  fibR(i)
  tf = time.clock()
  time_list.append(tf-ti)
  n_list.append(i)
tend = time.clock()

x = np.array([i for i in range(n)])
y = x
plt.plot(x,y, label="y=n")
plt.plot(n_list,time_list, label="time complexity")
title = "Recursive Fibonacci [0,50] time complexity"
plt.title(title)
plt.xlabel("n")
plt.ylabel("seconds")
plt.legend()
plt.show

print("Run for:",(tend-tstart)/60,"minutes")
```

```
x = np.array([i for i in range(n)])
y = x*10
plt.plot(x,y, label="y=10*n")
plt.plot(n_list,time_list, label="time complexity")
title = "Recursive Fibonacci [0,50] time complexity"
plt.title(title)
plt.xlabel("n")
plt.ylabel("seconds")
plt.legend()
plt.show
```

```
time_list1 = []
n_list1 = []
n=500
for i in range(n):
  ti = time.clock()
  fib(i)
  tf = time.clock()
  time_list1.append(tf-ti)
  n_list1.append(i)

x = np.array([i for i in range(n)])
y = x
plt.plot(x,y, label="n")
plt.plot(n_list1,time_list1, label="time complexity")
title = "Iterative Fibonacci time complexity"
plt.title(title)
plt.xlabel("n")
plt.ylabel("seconds")
plt.legend()
plt.show
```
